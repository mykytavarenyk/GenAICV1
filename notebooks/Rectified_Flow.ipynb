```python
import torch
import pytorch_lightning as pl
from pytorch_lightning.loggers import TensorBoardLogger
import torchvision
import matplotlib.pyplot as plt

from models.RectifiedFlow import (
    RectifiedFlowDiffusion,
    LatentRectifiedFlowDiffusion,
    MNISTDataModule
)

def train_pixel_diffusion(max_epochs=30, batch_size=64, logdir='lightning_logs', callbacks=None):
    data_module = MNISTDataModule(batch_size=batch_size)
    model = RectifiedFlowDiffusion(batch_size=batch_size)
    logger = TensorBoardLogger(logdir, name="pixel_diffusion")
    if callbacks is None:
        callbacks = [pl.callbacks.ModelCheckpoint(monitor='val_loss')]
    trainer = pl.Trainer(
        max_epochs=max_epochs,
        accelerator='auto',
        logger=logger,
        callbacks=callbacks,
        enable_progress_bar=True
    )
    trainer.fit(model, data_module)
    return model

def train_latent_diffusion(max_epochs=50, batch_size=64, latent_dim=128, logdir='lightning_logs', callbacks=None):
    data_module = MNISTDataModule(batch_size=batch_size)
    model = LatentRectifiedFlowDiffusion(
        batch_size=batch_size,
        latent_dim=latent_dim,
        img_size=28,
        lr=1e-4,
        kl_weight=0.001,
        scheduler_gamma=0.95
    )
    logger = TensorBoardLogger(logdir, name="latent_diffusion")
    if callbacks is None:
        callbacks = [pl.callbacks.ModelCheckpoint(monitor='val_loss')]
    trainer = pl.Trainer(
        max_epochs=max_epochs,
        accelerator='auto',
        logger=logger,
        callbacks=callbacks,
        enable_progress_bar=True
    )
    trainer.fit(model, data_module)
    return model

def main():
    pl.seed_everything(42)

    early_stop_callback = pl.callbacks.EarlyStopping(
        monitor='val_loss',
        patience=5,
        verbose=True,
        mode='min'
    )
    checkpoint_callback = pl.callbacks.ModelCheckpoint(
        monitor='val_loss',
        dirpath='checkpoints/',
        filename='diffusion-{epoch:02d}-{val_loss:.2f}',
        save_top_k=2,
        mode='min'
    )
    lr_monitor = pl.callbacks.LearningRateMonitor(logging_interval='epoch')
    callbacks = [checkpoint_callback, early_stop_callback, lr_monitor]

    print("Training pixel-based diffusion model...")
    pixel_model = train_pixel_diffusion(
        max_epochs=30,
        batch_size=64,
        callbacks=callbacks
    )

    print("Training latent-based diffusion model...")
    latent_model = train_latent_diffusion(
        max_epochs=50,
        batch_size=64,
        latent_dim=256,
        callbacks=callbacks
    )

    print("Generating samples from pixel-based diffusion model...")
    pixel_samples = pixel_model.generate_samples(num_samples=16)
    print("Generating samples from latent-based diffusion model...")
    latent_samples = latent_model.generate_samples(num_samples=16)


    fig, axes = plt.subplots(4, 8, figsize=(20, 10))
    for i in range(16):
        row, col = i // 4, i % 4
        axes[row, col].imshow(pixel_samples[i, 0].cpu().numpy(), cmap='gray')
        axes[row, col].set_title('Pixel-based')
        axes[row, col].axis('off')
    # Right side: Latent-based samples
    for i in range(16):
        row, col = i // 4, i % 4 + 4
        axes[row, col].imshow(latent_samples[i, 0].cpu().numpy(), cmap='gray')
        axes[row, col].set_title('Latent-based')
        axes[row, col].axis('off')

    plt.tight_layout()
    plt.savefig('comparison.png')
    plt.show()

if __name__ == "__main__":
    main()
```
